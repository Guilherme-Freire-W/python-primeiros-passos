```python
#Cadastro de notas do aluno
print("Digite as notas dos alunos: ")
a = float(input("Nota 1: "))
b = float(input("Nota 2: "))
c = float(input("Nota 3: "))
d = float(input("Nota 4: "))
notas = [a,b,c,d]
print("As notas inseridas são:")
print(notas)

#Calculo da média. Função para calcular a média
def calculo_media(notas):
 total = sum(notas)
 media = total / len(notas)
 return media
media = calculo_media(notas)

# Função para arrendodar media
arredondarMedia = lambda media: round(media,2)
mediaArredondada = arredondarMedia(media)


#Determinação da situação
print("A media mínima é 7.00.")
print("A média do aluno é:")
print(mediaArredondada)
if mediaArredondada >= 7.00:  #Se a média for maior ou igual a 7.00, o aluno está aprovado
    print("Situação do aluno: Aprovado.\nParabéns!")
elif mediaArredondada >= 5.00 and mediaArredondada <= 7:
  #Se a média for maior ou igual a 5.00 e menor ou igual a 7.00, o aluno está automaticamente em recuperação.
    print("Situação do aluno: Recuperação. Procure a secretaria para mais informações sobre as provas de recuperação")
else:
    print("Situação do aluno: Reprovado.")
    # Caso a média seja menor que 5.00, o aluno está reprovado.


#Relatório Final: a aplicação aponta as notas inseridas, a média e a situação do aluno, que varia conforme seu desempenho no trimestre.

```
